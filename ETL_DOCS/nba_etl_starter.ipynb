```python
import pandas as pd
```

```python
#read in data from 2 data sources
shot_types = pd.read_csv('nba_shot_types.csv')
season_data = pd.read_csv('Seasons_Stats.csv')
```

```python
season_data.columns
```

```python
#view columns
season_data.columns
shot_types.columns

```

```python
#groupby so that each player only has 1 row/ season
#name, year = unique identifier shared between both tables
shot_types_grouped = shot_types.groupby(['PLAYER', 'YEAR']).mean()
```

```python
shot_types_grouped.head()
```

```python
#this table should not need to be formated, but just in case. 
season_data_grouped = season_data.groupby(['Player', 'Year']).mean()
```

```python
season_data_grouped.head()
```

```python
#write both new data files to_csv, so they can be uploaded to pg admin
season_data_grouped.to_csv('season_data_grouped.csv')
```
